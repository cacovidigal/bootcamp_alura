Adicionar coluna mostrando região de cada estado 

```python
usaveis_sem_total.sort_values('2007/Out', ascending= False)
usaveis_sem_total.head()
```

```python
ordenar_por_total = usaveis_sem_total.sort_values('2007/Out', ascending= False)

ordenar_por_total.head()
```

```python
ordenar_por_total.head(5).T.plot(figsize = (10,6))
```

```python
colunas_interessadas = ordenar_por_total.columns[6:]
ordenar_por_total = ordenar_por_total[colunas_interessadas]
```

```python
ordenar_por_total.head(6).T.plot(figsize = (10,6))
plt.legend(bbox_to_anchor=(1.05,1),loc="best")
plt.title("Valor Gasto nos Procedimentos Hospitalares de 2008 a 2019")
plt.show()
```

```python
ordenar_por_total = ordenar_por_total / 1000000
ordenar_por_total.head(5).T.plot(figsize=(20,6))
plt.legend(bbox_to_anchor=(1.05,1),loc="best")
plt.title("Valor Gasto nos Procedimentos Hospitalares de 2008 a 2019")

plt.show()
```
